# Sample Mean and Sample Variance
Generate 10 deviates from an exponential distribution of a given mean and compute the Sample Mean and Sample Variance. Compare them with the real mean and variance. Then increase the sample to 100, 1000, .., 1,000,000. Observe the difference between the real mean and variance and the sample mean and variance.  
Repeat the experiment for a Pareto deviates of the same mean. Discuss differences. Use your generated data of random deviates to compare between the exponential and Pareto densities and the equivalent histograms. Use small enough ranges to achieve good fit.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Exponential Distribution

```python
from scipy.stats import expon

sample_num = [10, 100, 1000, 10000, 100000, 1000000]
mean, var = expon.stats(scale = 5, moments = 'mv')
means = []
variance = []

for i in range(len(sample_num)):
    expon_rv = expon.rvs(size = sample_num[i], scale = mean)
    means.append(np.mean(expon_rv))
    variance.append(np.var(expon_rv))

plt.figure(1)
plt.axes(xscale = "log")
plt.title("Means")
plt.xlabel('Sample Numbers')
plt.ylabel('Means')
plt.plot(sample_num, means)
plt.plot(sample_num, [mean,mean,mean,mean,mean,mean])


plt.figure(2)
plt.axes(xscale = "log")
plt.title("Variance")
plt.xlabel('Sample Numbers')
plt.ylabel('Variance')
plt.plot(sample_num, variance)
plt.plot(sample_num, [var,var,var,var,var,var])
```

The figures above show the mean and variance of the **Exponential Distribution** with a given mean of 5. As the sample size increased, the mean and variance of the generated random variables are more closed to the given mean and variance. When the sample size is 1,000, the differences are quite small.

## Pareto Distribution

```python
from scipy.stats import pareto

sample_num = [10, 100, 1000, 10000, 100000, 1000000]
b = 2.62
means = []
variance = []
mean, var = pareto.stats(b, moments='mv')

for i in range(len(sample_num)):
    pareto_rv = pareto.rvs(b, size = sample_num[i])
    means.append(np.mean(pareto_rv))
    variance.append(np.var(pareto_rv))

plt.figure(1)
plt.axes(xscale = "log")
plt.title("Means")
plt.xlabel('Sample Numbers')
plt.ylabel('Means')
plt.plot(sample_num, means)
plt.plot(sample_num, [mean,mean,mean,mean,mean,mean])


plt.figure(2)
plt.axes(xscale = "log")
plt.title("Variance")
plt.xlabel('Sample Numbers')
plt.ylabel('Variance')
plt.plot(sample_num, variance)
plt.plot(sample_num, [var,var,var,var,var,var])
```

The figures above show the mean and variance of the Pareto Distribution with a given mean of about 1.6. The conclusion is the same: with a larger sample size, the mean and variance of the random variables are more accurate.
